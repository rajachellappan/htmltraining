2. Use the below methods to normalize the following data: [10, 20, 35, 70, 100]:
  1. min-max normalization with min=0 and max=1.
  2. z-score normalization
  3. normalization with decimal scaling.

  [0.5 marks out of 5]

**Answer:**




2.1   min-max normalization with min=0 and max=1
      
      Min-max normalization = (value - min) / (max - min)
      data = [10, 20, 35, 70, 100]
      Result = [0, 0.11, 0.27, 0.67, 1]


	   
  
  2.2   z-score normalization

        z = (x – μ) / σ

        where μ - mean
              σ - standard deviation
        
        data = [10, 20, 35, 70, 100]
           μ = (10 + 20 + 35 + 70 + 100)/ 5 = 47
           σ = ((10-47)^2 + (20-47)^2 + (35-47)^2 + (70-47)^2 +(100-47)^2)/5 =  33.406
	
	    Applying z = (x – 47) / 33.406 for each value of the data

	    Result = [-1.107, -0.808, -0.359, 0.688, 1.586]




2.3   normalization with decimal scaling

      x = x/10^j
      data = [10, 20, 35, 70, 100]
      Result = [0.01, 0.02, 0.035, 0.07, 0.1]

5. Load the CSV file BL_books.csv, which includes records of books hosted at the British Library. Inspect the data and describe any issues observed with the data. Remove fields that refer to internal processes at the British Library and do not describe the books themselves (Corporate Author, Corporate Contributors, Issuance type, Former owner, Shelfmarks, Engraver). Clean data in column Date of Publication, as to remove the extra dates in square brackets (e.g. 1879 [1878] -> 1879) and convert date ranges to their start date, wherever present (e.g. 1860-63 -> 1860). Provide the code for data cleaning and provide evidence that your data cleaning methods have worked (e.g. by displaying a portion of the cleaned dataset). [0.5 marks out of 5]

**Answer:**

5. 
   Loading the data and retrieving the information:
image.png

Edition Statement, Place of Publication, Publisher, Author, Contributors should be categorized.

Date of Publication should be in datetime or int64.

Code:


````
import pandas as pd
#loading the csv file to dataframe
df = pd.read_csv('./BL-books.csv') 
#reading the file
pd.read_csv 

print('Before Cleanup:\n\n')
display(df[500:510])
#Dropping all internal processes columns
df.drop(['Corporate Author', 'Corporate Contributors', 'Issuance type',
      'Former owner', 'Shelfmarks', 'Engraver'], inplace=True, axis=1) 

#cleaning up the Date of Publication using regex
df['Date of Publication'] = df['Date of Publication'].str.extract(r'^(\d{4})') 
#converting the Datetype to numeric 
df['Date of Publication'] = pd.to_numeric(df['Date of Publication'],errors='coerce')
#converting NaN to '0'
df['Date of Publication'] = df['Date of Publication'].fillna(0) 
#converting NaN to more understandable
df['Edition Statement'] = df['Edition Statement'].fillna('No Edition')  
#Changing the data type to more relevant datatypes
df = df.astype({'Edition Statement':'category', 'Place of Publication':'category', 
                'Publisher':'category', 'Author':'category', 'Contributors':'category',
                'Date of Publication':'int64'}) 
#displaying data from 500 to 510
print('\n\n\n\n\nAfter Cleanup:\n\n')
df[500:510]
````

6. Load the CSV file country-income.csv which includes both numerical and categorical values. Perform data cleaning in order to replace any NaN values with the mean of the value for a given field. Then replace any categorical labels with numerical labels. Display the resulting dataset. You can use the sklearn.impute and sklearn.preprocessing packages to asssist you. [0.5 marks out of 5]

**Answer:** 

6.

````
import pandas as pd
from sklearn import preprocessing
df = pd.read_csv('./country-income.csv')
pd.read_csv

print('Before:')
display(df)

#Calculating the Mean of Age
meanOfAge = df['Age'].mean()

#Calculating the Mean of Income
meanOfIncome = df['Income'].mean()

#Replacing NaN values in 'Age' to Mean of Age
df['Age'] = df['Age'].fillna(meanOfAge)

#Replacing NaN values in 'Income' to Mean of Income
df['Income'] = df['Income'].fillna(meanOfIncome)

#Converting values in 'Online Shopper' from 'Yes' and 'No' to Boolean
df['Online Shopper']=df['Online Shopper'].map({'No':0, 'Yes':1})

#Changing all datetypes to the respective datatypes
df = df.astype({'Region':'category', 'Age':'int64','Income':'int64','Online Shopper':'bool'})


#Label Encoder to Label 'Region'
label_encoder = preprocessing.LabelEncoder()   
df['Region']= label_encoder.fit_transform(df['Region']) 
df['Region'].unique()

print("\n\nAfter:")
df
````

7. Load the CSV file shoesize.csv, which includes measurements of shoe size and height (in inches) for 408 subjects, both female and male. Plot the scatterplots of shoe size versus height for female and male subjects separately. Compute the Pearson's correlation coefficient of shoe size versus height for female and male subjects separately. What can be inferred by the scatterplots and computed correlation coefficients? You can implement your own formulation of the correlation coefficient or use the scipy.stats package to assist you. [1 mark out of 5]

**Answer:** 

7.

````
import pandas as pd
import scipy.stats
df = pd.read_csv('./shoesize.csv')
pd.read_csv

#Scatterplot for 'Female Shoe Size vs Height'
import matplotlib.pyplot as plt
df1 = df[df['Gender']=='F']
x1 = df1['Size']
y1 = df1['Height']

plt.scatter(x1, y1)
plt.title('Female - Shoe Size vs Height')
plt.xlabel('Shoe Size')
plt.ylabel('Height')
plt.show()

#Scatterplot for 'Male Shoe Size vs Height'
print('\n\n')
import matplotlib.pyplot as plt
df2 = df[df['Gender']=='M']
x2 = df2['Size']
y2 = df2['Height']

plt.scatter(x2, y2)
plt.title('Male - Shoe Size vs Height')
plt.xlabel('Shoe Size')
plt.ylabel('Height')
plt.show()
print('\n\n')
print('Pearson\'s Correlation Coefficient and p value for female = ',scipy.stats.pearsonr(x1, y1))
print('Pearson\'s Correlation Coefficient and p value for male = ',scipy.stats.pearsonr(x2, y2))
````

Analysis: 

Both for female and male the Correlation Coefficient is strong positive which means with increase in shoe size the height also increases. Coming to the p value, both female and male have p value lower than 0.05 which shows evidence for null hypothesis and very low percent of 5% chances of null is right.

8. Using the breast cancer dataset from section 1 of this notebook, perform Principal Component Analysis with 2 components. Compute the explained variance ratio for each component, and plot the scatterplot of all samples along the two principal components, color-coded according to the "Class" column (this column should not be used in the PCA analysis). Ensure that your data is normalized prior to performing PCA. What insights can you obtain by the explained variance ratio of each component, and by viewing the scatterplot of the principal components? [1 mark out of 5]


````
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data', header=None)
data.columns = ['Sample code', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape',
                'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin',
                'Normal Nucleoli', 'Mitoses','Class']
dataClass = data
data = data.drop(['Sample code','Class'], axis=1)

data['Bare Nuclei'] = pd.to_numeric(data['Bare Nuclei'],errors='coerce')

data =(data - data.mean())/data.std(ddof=0)
data = data.fillna(0)
data


#PCA
features =['Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape',
                'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin',
                'Normal Nucleoli', 'Mitoses']
# Separating out the features
x = data.loc[:, features].values
# Separating out the Class
y = dataClass.loc[:,['Class']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)

numComponents = 2
pca = PCA(n_components=numComponents)
#pca.fit(data)
principalComponents = pca.fit_transform(x)
projected = pd.DataFrame(data = principalComponents
             , columns = ['pc1', 'pc2'])

projectedResult = pd.concat([projected, dataClass['Class']], axis = 1)
display(projectedResult)

x = projectedResult['pc1']
y = projectedResult['pc2']



plt.plot(x,y,'o', color="blue");


plt.title('PC1 vs PC2 Scatterplot')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.show()

````
